```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt  
%matplotlib inline
```

```python
nsa = pd.read_csv('./Dataset/noshowappointments-kagglev2-may-2016.csv')
```

```python
Attendance = nsa.groupby('No-show')['Scholarship', 'Hipertension', 'Diabetes',
                                    'Alcoholism', 'Handcap', 'SMS_received'].sum()
Attendance
```

```python
Attendance.T
```

```python
Attendance.T.reset_index()
```

```python
Attendance.T.plot(kind = 'bar')
```

```python
Scholarship = nsa.groupby('Scholarship')['Hipertension', 'Diabetes',
                            'Alcoholism', 'Handcap', 'SMS_received'].sum()
Scholarship
```

```python
Scholarship.T
```

```python
Scholarship.T.reset_index()
```

```python
Scholarship.T.plot(kind = 'bar')
```

```python
Gen = nsa.groupby('Gender')['Scholarship', 'Hipertension', 'Diabetes',
                            'Alcoholism', 'Handcap', 'SMS_received'].sum()
Gen
```

```python
Gen.T
```

```python
Gen.T.reset_index()
```

```python
Gen.T.plot(kind = 'bar')
```

```python
SMS = nsa.groupby('SMS_received')['Scholarship'].sum()
SMS
```

```python
SMS.T
```

```python
SMS.T.reset_index()
```

```python
SMS.T.plot(kind = 'bar')
```
